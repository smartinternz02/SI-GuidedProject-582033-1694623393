**1) Univariate Analysis- Line-Graph, Scatter plot, Histogram and Distplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.bar(df.index,df['Price'])
plt.show()
```

```python
plt.scatter(df.index,df['lot area'])
plt.show()
```

```python
sns.distplot(df['Number of schools nearby'],kde=False)
plt.show()
```

```python
plt.hist(df['condition of the house'])
plt.show()
```

**2. Bivariate Analysis**-

```python
sns.jointplot(x=df['Distance from the airport'],y=df['Price'])
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.scatter(df['Distance from the airport'],df['Price'])
plt.xlabel('Distance from the airport')
plt.ylabel('Price')
plt.show()
```

**3. Multivariate Analysis**

```python
a=df.corr()
sns.heatmap(a,annot=False)
```

```python
sns.pairplot(df,vars=['living area','lot area','number of floors','Distance from the airport','Price'])
```
